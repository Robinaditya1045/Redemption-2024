```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Movie-Dataset-Latest.csv')
df.head()
```

```python
df.shape
```

```python
top_20 = df.sort_values(by='popularity', ascending=False).head(20)
```

```python
title = top_20['title']
# popularity = df['popularity']
popularity = top_20['popularity']
```

```python
plt.figure(figsize=(10, 6))
plt.barh(title, popularity, color='red')
plt.xlabel('Popularity')
plt.title('Top 20 Movies by Popularity')
plt.show()
```
